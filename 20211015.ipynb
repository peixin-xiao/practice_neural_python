```python
data2=df.iloc[range(1,48),1]
data2 = np.array(data2)
data2 = np.squeeze(data2)
data4 = np.zeros((47,2),dtype=float)
for flag  in range(0,46):
    if data2[flag]==0:
        data4[flag,1]=1
    else:
        data4[flag,0]=1      
print(data2.shape)
print(data4.shape)
data4
```

```python
class LogisticRegression(object):
    
    #Multiclass Logistic Regression
    
    def __init__(self, input_dim, output_dim):
        self.input_dim = input_dim
        self.W = np.random.normal(size=(input_dim, output_dim))
        self.b = np.zeros(output_dim)
        
    def __call__(self, x):
        return self.forward(x)
    
    def forward(self,x):
        return softmax(np.matmul(x, self.W) + self.b)
    
    def compute_gradients(self, x, t):
        y = self.forward(x)
        delta = y - t
        dW = np.matmul(x.T, delta)
        db = np.matmul(np.ones(x.shape[0]), delta)
        return dW, db
    
```

```python
data3=data.iloc[range(1,48),range(1,9)]
data3=np.array(data3)
#data3 = data3.astype(np.float32)
print(data3.shape)
data3
```

```python
#Softmax Fuction

def softmax(x):
    c = x.max()
    return (np.exp(x)*np.exp(-c) )/( np.sum(np.exp(x), axis=1, keepdims=True)*np.exp(-c))



M = 8
K = 2

x = data3
t = data4
#2.Model


model = LogisticRegression(input_dim=M, output_dim=K)

#3.Train Model

def compute_loss(t,y):
    return (-t * np.log(y)).sum(axis=1).mean()

def train_step(x, t):
    dW, db = model.compute_gradients(x, t)
    model.W = model.W - 0.1 * dW
    model.b = model.b - 0.1 * db
    loss = compute_loss(t, model(x))
    return loss

epochs = 10
batch_size = 47
n_batches = x.shape[0] // batch_size

for epoch in range(epochs):
    train_loss = 0
    x_, t_ = shuffle(x, t)

    for n_batch in range(n_batches):
        start = n_batch * batch_size
        end = start + batch_size

        train_loss += train_step(x_[start:end], t_[start:end])

    if epoch % 10 == 0 or epoch == epochs - 1:
        print('epoch: {}, loss: {:.3f}'.format(epoch+1, train_loss))

#4.Evaluate Model

```
